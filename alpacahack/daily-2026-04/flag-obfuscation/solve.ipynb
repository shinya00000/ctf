```python
import ipaddress
import re
import os
from google.colab import drive

def solve():
    drive.mount('/content/drive')

    file_path = '/content/drive/MyDrive/CTF/data.h'

    if not os.path.exists(file_path):
        print(f"Error: {file_path} が見つかりません．パスを確認してください．")
        return

    # data.h を読み込む
    with open(file_path, "r", encoding="utf-8") as f:
        content = f.read()

    # 正規表現でダブルクォーテーションで囲まれたIPv6文字列を抽出
    matches = re.findall(r'"([^"]+)"', content)

    output_data = b""
    for ip_str in matches:
        try:
            # IPv6文字列を16バイトのバイト列に変換
            ip = ipaddress.IPv6Address(ip_str)
            output_data += ip.packed
        except ValueError:
            continue

    # 復元したデータを保存
    output_path = "restored.exe"
    with open(output_path, "wb") as f:
        f.write(output_data)

    print(f"[+] 復元が完了しました：{output_path}")

if __name__ == "__main__":
    solve()
```
